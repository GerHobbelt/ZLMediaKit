```python
import requests
```

```python
secret="YOUR_SECRET"
```

```python
def reg(src,out):
    r = requests.get(f'http://127.0.0.1:8080/index/api/addFFmpegSource?src_url={src}&dst_url=rtmp://127.0.0.1/testing/web{out}&timeout_ms=100000&ffmpeg_cmd_key=ffmpeg.cmd&secret={secret}')
    return r.json()

def get():
    r = requests.get(f'http://127.0.0.1:8080/index/api/getMediaList?secret={secret}')
    return r.json()
```

```python
reg('rtsp://wowzaec2demo.streamlock.net/vod/mp4:BigBuckBunny_115k.mp4',1)
a = get()
```

```python
# test on reg same output url but different source
reg('towercam04.rep-am.com:8084/mjpg/video.mjpg',1)
b = get()
```

```python
#get originUrl for each stream   
def geturl(a):
    a_stream = []
    for i in a['data']:
        a_stream.append((i['originUrl'],i['schema'],i['stream']))
    return a_stream

a_stream = geturl(a)
b_stream = geturl(b)

def check(a_stream,b_stream):
    #check if the same output url has different source
    for i,st in enumerate(a_stream):
        if st != b_stream[i]:
            print('same output url has updated')
        else:
            print('same output url is the same') # it means new source will not replace the old one
check(a_stream,b_stream)
```
